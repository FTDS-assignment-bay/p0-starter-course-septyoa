```python
# Berikut adalah library yang dipanggil untuk pengerjaan NGC ini
import pandas as pd
import time
from selenium import webdriver
driver = webdriver.Edge()
from bs4 import BeautifulSoup
```

## Task 1 WebScrap from https://coinmarketcap.com/
Task:
1. Scrape up to 100 coins
2. Including Price, Market Cap, Volume, and Coin Supply
3. Save into Pandas DataFrame and then save into excel or csv file

```python
# Membuat daftar list kosong yang kemudian akan menjadi kolom tabel
coinname = []
price = []
marketcap = []
volume = []
supply = []
```

```python
# Digunakan for loop agar bisa scrape data dari berbagai page
for page in range (1,11):
    url1=f"https://coinmarketcap.com/?page={page}"
    driver.get(url1)
    # code untuk aktivasi scroll, jangan lupa panggil modul time
    scroll_pause_time= 1 # menentukan waktu jeda sebelum scroll
    screen_height = driver.execute_script("return window.screen.height;")
    i = 1
    while True:
        driver.execute_script(f"window.scrollTo(0, {screen_height * i});") #code untuk mulai scroll
        i += 1
        time.sleep(scroll_pause_time)
        scroll_height = driver.execute_script("return document.body.scrollHeight;")
        if screen_height * i > scroll_height:
            break
    html1 = driver.page_source
    coin = BeautifulSoup(html1, "html.parser")
    cresult1 = coin.find("tbody")
    for point in cresult1:
        CoinName = point.find("div", {"class":"sc-4c05d6ef-0 bLqliP"})
        if CoinName!= None: coinname.append(CoinName.get_text())
        else : coinname.append(None)
    for point in cresult1:
        Price= point.find("div", {"class":"sc-b3fc6b7-0"})
        if Price != None: price.append(Price.get_text())
        else : price.append(None)
    for point in cresult1:
        MarketCap= point.find("span", {"class":"sc-11478e5d-1 jfwGHx"})
        if MarketCap!=None: marketcap.append(MarketCap.get_text())
        else : marketcap.append(None)
    for point in cresult1:
        Volume= point.find("p", {"class":"sc-71024e3e-0 bbHOdE font_weight_500"})
        if Volume!=None: volume.append(Volume.get_text())
        else : volume.append(None)
    for point in cresult1:
        Supply =point.find("div", {"class":"sc-4c05d6ef-0 sc-552ef8b1-1 dlQYLv klYCgG"})
        if Supply!= None: supply.append(Supply.get_text())
        else : supply.append(None)
```

```python
# Pembuatan DataFrme Pandas dan convert list tadi menjadi kolom dalam tabel
dfc = pd.DataFrame()
dfc['coin_name'] = coinname
dfc['price'] = price
dfc['market_cap'] = marketcap
dfc['volume'] = volume
dfc['supply'] = supply
```

```python
# Pemanggilan DataFrame Coins
dfc
```

```python
# Save tabel coin to csv file dengan nama coins.csv
dfc.to_csv('coins.csv')
```

## Task 2: Webscraping information of 54 location in indonesia from https;//www.timeanddate.com/weather/indonesia
Information must include:
1. Temperature
2. Feels like
3. Wind Speed
4. Wind Direction
5. Humidity
6. Dew Point
7. Visibility
8. Probability of Rain
9. Amount of Rain

```python
# kita buat list nama kota
kota  = []
```

```python
# buat scrape nama kota untuk lebih lanjut scrape informasi detail pada langkah berikutnya
url2= "https://www.timeanddate.com/weather/indonesia"
driver.get(url2)

soup = BeautifulSoup(driver.page_source, "html.parser")
kresult = soup.find_all("td")
for city in kresult:
    nk = city.find("a")
    if nk != None:
        kota.append(nk.get_text())
    else : continue
```

```python
# cleaning data nama kota : mengganti spasi dengan "-", serta merubah nama kota yang tidak sesuai dengan url yang tersedia
kota1 = list(dict.fromkeys(kota))
kota1 = [Kota.replace(" ", "-") for Kota in kota]
for z in range(len(kota)):
    if kota1[z] == "Bandar-Lampung": kota1[z] = "tanjungkarang"
    if kota1[z] == "Ende": kota1[z] = "endeh"

#fungsi ini untuk memudahkan penambahan hasil scrape ke dalam list
def tambah (item,kolom):
    kolom.append(item.find("td").get_text())

#membuat list dari info yang akan diambil
temp = []
feel = []
wspeed = []
wdir = []
humid = []
dewp = []
vis = []
rprob = []
amt = []
```

```python
# membuat loop agar data diambil dari berbagai webpage
for Kota in kota1:
    url3 = f"https://www.timeanddate.com/weather/indonesia/{Kota}" 
    driver.get(url3)
    cuaca = BeautifulSoup(driver.page_source, "html.parser")
    tabel = cuaca.find("table", {"class":"zebra tb-wt tc sep"})
    body = tabel.find("tbody")
    result0 = body.find("tr")
    result1= result0.find_next("tr")
    result2 = result1.find_next("tr")
    result3 = result2.find_next("tr")
    result4 = result3.find_next("tr")
    result5 = result4.find_next("tr")
    result6 = result5.find_next("tr")
    result7 = result6.find_next("tr")
    result8 = result7.find_next("tr")
    result9 = result8.find_next("tr")
    result10 = result9.find_next("tr")
    tambah(result1,temp)
    tambah(result3,feel)
    tambah(result4,wspeed)
    tambah(result5,wdir)
    tambah(result6,humid)
    tambah(result7,dewp)
    tambah(result8,vis)
    tambah(result9,rprob)
    tambah(result10,amt)
```

```python
# membuat dataframe weather
dfw = pd.DataFrame()
dfw['Kota'] = kota
dfw['Temperature'] = temp
dfw['Feels Like'] = feel
dfw['Wind Speed'] = wspeed
dfw['Wind Direction'] = wdir
dfw['Humidity'] = humid 
dfw['Dew Point'] = dewp
dfw['Visibility'] = vis
dfw['Probabilty of Rain'] = rprob
dfw['Amount of Rain'] = amt
```

```python
# memanggil dataframe weather
dfw
```

```python
# save dataframe weather ke file csv dengan nama "weather.csv"
dfw.to_csv('weather.csv')
```
